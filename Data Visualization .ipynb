```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
years = [2010,2011,2012,2013,2014,2015]
yield_apples = [0.895,0.91,0.919,0.926,0.929,0.931]
plt.plot(years,yield_apples);
plt.title("Apple Yield")
plt.xlabel('year');
plt.ylabel('yield (tons per hectare)');
```

```python
days = [15,16,17,18,19,20,21,22,23]
rainfall_in_kutch = [10,20,3,5,23,15,18,19,2]
rainfall_in_ahmedabad = [25,30,40,8,15,20,30,26,10]
```

```python
plt.plot(days,rainfall_in_kutch,'o-r');
plt.plot(days,rainfall_in_ahmedabad,'x--b')
plt.legend(['Kutch','Ahmedabad'])
plt.xlabel("Dates (July 2021)")
plt.ylabel("Rainfall(in cms)")
plt.title("Rainfall In Kutch VS Rainfall In Ahmedabad")
plt.plot
sns.set_style("darkgrid")

```
